```python
import requests, json
from pprint import pprint

class PokeAPI():
    new_pokedek = []
    
    #What this class must be able to do:
    #1. Pull data from the PokeAPI
    #2. Add pokemon to an empty pokedex 
    #3. Show a new list of pokemon (could also be put into a run class)
    #4. Sort new list by category (type) within said list
    
    #1. Pull data from the PokeAPI
    @classmethod
    def poke_data(cls, pokemon_number):
        responses = requests.get(f'https://pokeapi.co/api/v2/pokemon/{pokemon_number}')
        data=responses.json()
        p_dict = {
            'abilities': data['abilities'][0]['ability']['name'],
            'name': data['name'],
            'type':data['types'][0]['type']['name'],
            'weight':data['weight']
        }
        #each time data is gotten, call the add_pokemon method
        cls.add_pokemon(p_dict)
        
    #2. Add pokemon to an empty pokedex 
    @classmethod
    def add_pokemon(cls, pokemon_data):
        cls.new_pokedek.append(pokemon_data)
    
    #3. Show a new list of pokemon (could also be put into a run class)
    @classmethod
    def show(cls):
        for pokemon in cls.new_pokedek:
            pprint(pokemon)
    
    #4. Sort new list by category (type) within said list
    @classmethod
    def sort_by(cls):
        #new dict where each type is a key, and each pokemon is a value of said key
        pokemon_dict = {}
        #for pokemon in our new list
        for pokemon in cls.new_pokedek:
            #if their type is NOT in the list, add the dict of type.
            if pokemon['type'] not in pokemon_dict:
                pokemon_dict[pokemon['type']]={}
            #update the type dict with the pokemons name it could contain and those abilities and weight as another dictionary within.
            pokemon_dict[pokemon['type']].update({
                pokemon['name']:
                {
                    'abilities': pokemon['abilities'],
                    'weight':pokemon['weight']
                }
            })
        pprint(pokemon_dict)
            

p = PokeAPI        
p.poke_data(1)
p.poke_data(2)
p.poke_data(3)
p.poke_data(4)
p.poke_data(5)
p.poke_data(6)
p.poke_data(7)
p.poke_data(8)
p.poke_data(9)
p.poke_data(10)
p.poke_data(11)
p.poke_data(12)
p.poke_data(13)
p.poke_data(14)
p.poke_data(15)
p.poke_data(16)
p.poke_data(17)
p.poke_data(18)
p.poke_data(19)
p.poke_data(20)
p.sort_by()
```
